### Problem 1. The FizzBuzz Test

A. The FizzBuzz test originated from children's division games and was an often administered interview question in industry programmer hires.

We take the text of the assignment from [here](https://wiki.c2.com/?FizzBuzzTest).

> "Write a program that prints the numbers from 1 to 100. But for multiples of three print “Fizz” instead of the number and for the multiples of five print “Buzz”. For numbers which are multiples of both three and five print “FizzBuzz”."



```python
# Define my variables

start = 1
stop = 101
```

```python
# Write a function for printing FizzBuzz

def fizzbuzz(start:int=1, stop:int=101):
  '''
  This function executes the FizzBuzz test.

  Optional keywords:
  start:  int, starting integer
  stop:   int, final integer + 1
  '''
  
  for i in range(start, stop):
    # First print FizzBuzz
    if i % 3 == 0 and i % 5 == 0:
      print("FizzBuzz")

    # Next print Fizz
    elif i % 3 == 0:
      print("Fizz")

    # Next print Buzz
    elif i % 5 == 0:
      print("Buzz")

    # Finally print everything else
    else:
      print(str(i))
```

**Answer:**

```python
fizzbuzz(start,stop)
```

```python
# This can be run in jupyter notebook using some notebook magic
%timeit fizzbuzz(start,stop)
```
